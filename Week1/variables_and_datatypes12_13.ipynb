# what is list?



```python
#list is  orderd and changeable collection 
fruits=["apple","orange","banana","mango"]
print(f"fruits list is  {fruits}")
```

# Accessing elements

```python
print(f"first element of the list is {fruits[0]}") #indexing starts from 0
print(f"last element of the list is {fruits[-1]}") #negative indexing starts from -1
```

# list methods

```python
fruits.append("grapes") # to add element at the end of the list
print(f"fruits list after adding grapes is {fruits}")
print("\n ","-"*50, "\n") 



fruits.insert(1,"grapes") # to add the element at specific 
print(f"fruits list after adding grapes at index 1 is {fruits}")
print("\n ","-"*50, "\n")



fruits.remove("mango") # to remove the element
print(f"fruits list after removing mango is {fruits}")
print("\n ","-"*50, "\n")



fruits.pop() # this will remove last element of the 
print(f"fruits list after removing last element is {fruits}")
print("\n ","-"*50, "\n")



fruits.sort() # to sort the list according to alphabetical order
print(f"fruits list after sorting is {fruits}")
```

# Looping Through list

```python
for fruits in fruits:
    print(fruits)#print fruits one by one using for loop
```

# what is tuple?

```python
# A tuple is a collection which is ordered and unchangeable. It is defined by using parentheses ()
colors=("red","orange","blue")
print(f"colors tuple is {colors}")
```

# Accessing tuple


```python
print(f"{colors[1:]}") # print the elements from index 1 to end of the tuple
```

# Tuple  properties

```python
#tuples are immutable (unchangeable)
colors[0]="yellow" # this will cause an error
```

# Tuple Methods

```python
numbers=(1,2,3,2,5,4)
print(f"how many times two repeat? \n{numbers.count(2)}") # to count how many times 2 repeat in the tuple

print(f"index of 5 is? \n{numbers.index(5)}") # to find the index of 5 in the tuple
```

# What is Set

```python
#A set is collection which is unordered and unindexed. It is defined by using curly braces {}
my_set={1,2,3,4,4,5}
print (my_set)# it will remove the duplicate element 4 and print only one 4 in the set
```

# sets methods

```python
my_set.add(7) # to add element in the set
print(f"my set after adding 7 is {my_set}")

my_set.remove(2) # to remove element from the set
print(f"my set after removing 2 is {my_set}")
```

# Sets Opetrations

```python
# taking two sets
A={3,2,4}
B={2,5,6,1}
print(f"{A.union(B)} is union set of {A} and {B}")
print(f"{A.intersection(B)} is intersection set of {A} and {B}")
print(f"{A.difference(B)} is difference set of {A} and {B}")
```

# What is dictionary

```python
# A set is a collection which is unordered and unindexed, key value pairs . It is defined by using curly braces {}.
student={
    "name":"Attia Ubaid",
    "age":20,
    "course":"BSCS"}
print(f"Student data, {student}")

```

# Access Values


```python
print(student["name"])
print(student.get("age"))
```

# Dictionary methods

```python
student["age"]=21 # to change the value of age key
student["city"]="Karachi" # to add new key value pair in the dictionary
student.pop("course")#Remove
print(student)
```

# Looping dictionary

```python
for key ,values in student.items():
    print(key," ", values)
```
