# Topic 52
## Numpy indexing slicing and strides

```python
import numpy as np
arr=np.array([[10,20,30],
            [40,50,60],
            [70,80,90]])
```

```python
print(arr[0,1])# 20 will be printed
print(arr[2,2])# 90 will bw printed
```

```python
# Negative indexing
print(arr[-1,-2])
print(arr[-3,-3])
```

```python
#portion of array
#general form :arr[start,stop,step]
print(arr[0:2,1:3])
```

```python
print(arr[::2,::3])
```

```python
slice_arr=arr[0,:2]#view (not copy)
slice_arr[0]=99
```

```python
print(arr[0])# original array change 
```

```python
copy_arr=arr[1].copy()
copy_arr[0]=555
```

```python
print(copy_arr[0])
```

```python
print(arr[0])
```

### Numpy random numbers,reproducibilityand performance tips

```python
import numpy as np 
import time
```

```python
# Random number produces numbers randomly
#see_function use to generate random numbers from a specific distribution
rng=np.random.default_rng(seed=43)
flips=rng.integers(0,2,size=100000)
probs_heads=np.mean(flips)
print(f"estiamted probability of heads : {probs_heads}")
```

```python
# python loop 
start=time.time()

square=[i**2 for i in range(100000)]
end=time.time()
print(f"Loop time: {end-start}")
```

```python
#Numpy vectorized operations
start=time.time()

arr=np.arange(100000)
sqr_np=arr**2
end=time.time()
print(f"Numpy Vectorized time : {end-start}")
```
