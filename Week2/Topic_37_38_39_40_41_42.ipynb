# Lecture 37
# Method Overridding

```python
class Animal():
    def speak(self):
        print("Some Sound")
class Dog(Animal):
    def speak(self):
        print("woof woof!")
dog1=Dog()
dog1.speak()#inherent class method override by parent class
```

# Lecture 38
### polymorphism in python overridding and overloading 

```python
class Dog():
    def speak(self):
        print("woof!")
class Cat():
    def speak(self):
        print("Meow")
```

```python
for pet in [Dog(),Cat()]:
    pet.speak()

    
```

```python
class Vhicle():
    def move(self):
        print("VHICLE is moving")
class Car(Vhicle):
    def move(self):
        print("car is driving")
car1=Car()
car1.move()# this concept is known as overriding
```

```python
class Calculator():
    def add(self,a,b=0,c=0):
        return a+b+c
calc1=Calculator()
print(calc1.add(5))
print(calc1.add(5,4))
print(calc1.add(5,4,3))# That concept is known as overloading

```

## python special method

```python
class Book():
    def __init__(self,title,author):
        self.title=title
        self.author=author
    def __str__(self):
        return f"{self.title} by {self.author}"
b=Book("1894 ML","Goorge orwell")
print(b)
```

```python
#Example 2
class playlist():
    def __init__(self,songs):
        self.songs=songs
    def __len__(self):
        return len(self.songs)
    def __add__(self,other):
        return playlist(self.songs+other.songs)
p1=playlist(["song1","song2"])
p2=playlist(["song1","song2","song3"])
print(len(p1))
print(len(p1+p2))
```

```python
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"{self.x},{self.y}"
    def __add__(self,other):
        if not isinstance(other,point):
            return NotImplemented
        return point(self.x+other.x,self.y+other.y)
p1=point(2,3)
p2=point(4,1)
print(p1+p2)
```
