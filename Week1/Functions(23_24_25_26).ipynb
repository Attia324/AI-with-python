# Topic 26 
# lambda function 


```python
add=lambda a,b:a+b
print(add(4,232))
```

```python
double_it=lambda x:x*2
print(double_it(4))
```

```python
#we can also use functions map() filter()
num=[1,2,3,4,5]
square=list(map(lambda x:x*x,num))
print(square)
```

```python
##  filter function
num=[1,2,3,4,5,6,7,8]
odd_number=list(filter(lambda x:x%2!=0,num))
print(odd_number)
```
