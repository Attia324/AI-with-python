# For loop 

```python
for  i in range (3): # dispaly 3 times
    print("Hello Dear")
```

```python
for i in range (1,11): 
    print(i)
```

```python
for i in "Python":
    print(i)
```

```python
i=4
rows=i
cols=i
matrix=[]
for i in range(rows):
    row=[]
    for c in range(cols):
        row.append(0)
    matrix.append(row)
for row in matrix:
    print(" ".join(map(str,row)))
```

A loop inside another loop is called nested loop
its is not best practise to use nested loop because  become complex to understand

```python
for i in range(3):
    print(f"Outer loop")
    for j in range(3):
        print(f"          Inner loop")
```

# Lecture 22
## looping  over data structure (dictionaries)


```python
students={"name":"Ali",
          "age":20,
          "grade":"A"}
```

```python
for key in students:
    print(key)
```

```python
#printing value
for key in students:
    print(students[key])
```

```python
# for both key and value
for key,value in students.items():
    print(f"{key}:{value}")
```

```python
# dictionaries in list
students=[
    {"name":"Ali",
     "age":20,
     "grade":"A"},
    {"name":"Sara",
     "age":22,
     "grade":"B"},
    {"name":"Omar",
     "age":21,
     "grade":"A"}
]
for  student in students:
    print(f"name : {student.get('name')}, age : {student.get('age')}, grade : {student.get('grade')}  ")
```
