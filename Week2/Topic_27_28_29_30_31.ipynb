# Lecture 29
we will practise the concept of class and objects in this lecture

```python
class Car:
    def __init__(self,color):
        self.color=color
    def start(self):
        print(f"{self.color} car is started")

```

```python
#class is created now we will create an object of that class
my_car=Car("red")
my_car.start()
my_brothers_car=Car("blue")
my_brothers_car.start()
```

# lecture 30 :    Defining class and creating object


```python
#creating class
class Car:
    def drive(self):
        print("The car is moving")
# creating object of that class
car1=Car()
car1.drive()
```

```python
class Car:
    color="red"
    def drive(self):
        print(f"{self.color} car is moving" )
    def set_color(self,new_color):
        self.color=new_color
#Creating object
car2=Car()
car2.drive()
```

```python
car2.set_color("green")
car2.drive()
```

# lecture 31:      The __init__ constructor method

```python
class Dog:
    def __init__(self):
        self.name="Buddy"
    def bark(self):
        print(f"{self.name} says Woof !")
        
```

```python
dog1=Dog()
dog1.bark()
```

```python
class Car:
    def __init__(self,brand,color):
        self.color=color
        self.brand=brand
    
```

```python
car1=Car("toyota","black")
print(car1.color)
print(car1.brand)
```

```python
car1.color="blue"
print(car1.color)
```

```python
car2=Car("Honda","White")
print(car2.brand)
```
