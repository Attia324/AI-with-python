# Nested if

```python
age =19
country="pakistan"
if age>=18:
    if country=="pakistan":
        print("You can vote in pakistan")
else :
        print("you cannot vote in pakistan")
```

```python
age=int(input("Enter your age: "))
if age>=18:
    country=input("Enter your Country: ").capitalize()
    if country=="Pakistan":
        print("You can vote in pakistan")
    else:
        print("you cannot vote in pakistan")
else:
    print("You are not eligible")
```

```python
age=int(input("Enter your age: "))
country=input("Enter your Country: ").capitalize()
if age>=18 and country=="Pakistan":
    print("Eligible")
else:
    print("Not eligible")
```
