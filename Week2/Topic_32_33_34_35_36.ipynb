# Instance variable and instance method

### Instance
#### an instance  refer to an object created form a class

```python
class Student():
    def __init__(self, name,age, grade):
        self.name=name
        self.age=age
        self.grade=grade
    def display_info(self):
        print(f"{self.name} is {self.grade} student and {self.age} year old.")
    def is_eligible(self):
        if self.age>=18:
            print(f"{self.name} is eligible")
        else:
            print("Not eligible")
```

# Lecture 33
### Accessing object Attributes and method


```python
#creating object
student1=Student("aysha",19,"12th")
student2=Student("zenea",5,"1th")
#Accessing attributes
print(student1.name)
print(student1.grade)
#Accessing methods
student1.display_info()
student2.is_eligible()
```

```python
class Stud:
    def __init__(self,name,grade):
        self.name=name
        self.__grade=grade # private atribute
    def get_grade(self):
        return self.__grade
student1=Stud("Ali","5th")
print(student1.get_grade())
```

# Lecture 35
# Getter And Setter 

```python
class Student:
    @property
    def grade(self):
        return self.__grade
    def __init__(self,name,grade):
        self.name=name
        self.__grade=grade
    def get_grade(self):
        return self.__grade
    def set_grade(self,grade):
        if 0<=grade<=grade:
            self.__grade=grade
        else:
            print("Invalid Grade")


```

```python
student1=Student("Ali",90)
student1.set_grade(95)
student1.get_grade()
```
