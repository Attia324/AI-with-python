### Handling Exception during file operation


```python
with open("missing data.txt","r") as file:
    data=read(file)
    print(data)# error because missing data .txt doesnot exist
```

```python
file_name="missing data.txt"
try:
    with open(file_name,"r") as file:
        data=read(file)
        print(data)
except FileNotFoundError:
    with open(file_name,"w") as fout:
        fout.write(f" I am Busy")
    with open(file_name,"r") as fin:
        print(fin.read())
```

```python
try:
    num=int(input('Please Enter the number'))
    print(10/num)
except ZeroDivisionError:
    print("You cannot divide with 0")
except ValueError:
    print('Thats not number')
except Exception as e:
    print(f" An unexpected error occur")
    
```
