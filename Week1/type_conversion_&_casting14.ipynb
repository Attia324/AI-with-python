# Type Conversion and Casting
we can convert one type of data into another type                                               
string into int vice versa                                        
float to int vice versa etc


```python
Age="54"
print(f"my age is {Age}, type of age is  {type(Age)}") # Age type is string
```

```python
#converting string to integer
print(f"My age is {int(Age)+1} and type is {type(int(Age))}") # Age type is integer
```

```python
hieght=4.3
print(f"height :{int(hieght)} \n  type:{type(int(hieght))}") # converting float to integer
```

```python
character='a'
print(f"character :{character} \n  type:{type(character)}") # character type is string
```
